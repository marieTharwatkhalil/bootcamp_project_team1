```python
import pandas as pd 
from textblob import TextBlob
```

```python
def label_reviews(reviews_labeled_df):
    """this function takes the unlabeled reviews dataframe 
    and set a label for each review 
    1 is positive 
    -1 is negative 
    0 is neutral"""
    for i in range(len(reviews_labeled_df)):
        review_sent_=TextBlob(reviews_labeled_df["Reviews"][i])
        review_sentiment= review_sent_.sentiment[0]

        if review_sentiment>0:
            reviews_labeled_df["Label"][i]=1
        if review_sentiment<0:
            reviews_labeled_df["Label"][i]=-1
        if review_sentiment==0:
            reviews_labeled_df["Label"][i]=0        
        
        if reviews_labeled_df["Reviews"][i]=='five stars' or '5 stars' or 'four stars'or '4 stars'or '3 stars' or 'three stars':
            reviews_labeled_df["Label"][i]=1
        print(i)
        print(reviews_labeled_df["Reviews"][i],reviews_labeled_df["Label"][i])
        
    return reviews_labeled_df
```

```python
data= pd.read_csv("Cleaned_reviews",index_col=False)
```

```python
#read the new labeled dataframe with new indexes 
reviews_labeled=pd.read_csv("reviews_labeled")
```

```python
reviews_with_label=label_reviews(reviews_labeled)
```
