```python
#reading the excel file
sales = pd.read_excel(r"data//original_data")
#text_lowercase("Full Text Reviews") 
sales
```

```python
#adding 'Full Text Reviews ' and 'Title of Reviews ' in one column
reviews__= np.concatenate([sales['Full Text Reviews '],sales['Title of Reviews ']])
reviews__
```

```python
#converting the reviews array into a dataframe 
reviews__DF=pd.DataFrame(reviews__)
```

```python
reviews__DF["Reviews"]=pd.DataFrame(reviews__)
#reviews__DF["reviews"]=reviews__DF
#reviews__DF.co
reviews__DF
```

```python
#dropping the column 0
reviews__DF.drop(labels=[0],axis=1,inplace=True)
```

```python
#the reviews dataframe is ready to  be cleaned 
reviews__DF
```

```python

#Remove NAN Values
#df_new = df[df['Reviews'].notnull()] 
df_not_null=reviews__DF[reviews__DF['Reviews'].notnull()]
df_new=df_not_null
```

```python
#Put all in lower case
df_new['Reviews'] = df_new['Reviews'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
#Remove symobls
df_new['Reviews']= df_new['Reviews'].str.replace('[^\w\s]','') #
df_new.head(10)
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
#Remove StopWords
df_new['Reviews']= df_new['Reviews'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df_new
```

```python
#the most frequent words 
freq = pd.Series(' '.join(df_new['Reviews']).split()).value_counts()[:10]
freq
```

```python
#saving the dataframe 
df_new.to_csv("Cleaned_reviews",index=False)
```
